# Now Let's read this HPLC data, and get some functional type info outta there

```r
require(dplyr, warn.conflicts = FALSE);
#require(ggplot2);
require(cowplot, warn.conflicts = FALSE); # for nicer plots, includes ggplot2
require(chron, warn.conflicts = FALSE);
require(ncdf4, warn.conflicts = FALSE);
```

```r
# open a NetCDF file
HPLC_Mote <- nc_open("HPLC_Mote.netcdf")

# print(nctdin)

HPLC_Mote_df <- data.frame("Num" = 1:HPLC_Mote$dim$unlimited$len)
for(i in 1:HPLC_Mote$nvars) {
    HPLC_Mote_df[paste(attributes(HPLC_Mote$var)$names[i])] <- ncvar_get(HPLC_Mote, attributes(HPLC_Mote$var)$names[i])
}
head(HPLC_Mote_df)
```

```r
# Write CSV in R
write.csv(HPLC_Mote_df, file = "HPLC_Mote_df.csv")
```

```r
# open a NetCDF file
HPLC_BBSR <- nc_open("HPLC_BBSR.netcdf")

# print(nctdin)

HPLC_BBSR_df <- data.frame("Num" = 1:HPLC_BBSR$dim$unlimited$len)
for(i in 1:HPLC_BBSR$nvars) {
    HPLC_BBSR_df[paste(attributes(HPLC_BBSR$var)$names[i])] <- ncvar_get(HPLC_BBSR, attributes(HPLC_BBSR$var)$names[i])
}
tail(HPLC_BBSR_df)
```

```r
# Write CSV in R
write.csv(HPLC_BBSR_df, file = "HPLC_BBSR_df.csv")
```

```r
# open a NetCDF file
HPLC_HPL <- nc_open("HPLC_HPL.netcdf")

# print(nctdin)

HPLC_HPL_df <- data.frame("Num" = 1:HPLC_HPL$dim$unlimited$len)
for(i in 1:HPLC_HPL$nvars) {
    HPLC_HPL_df[paste(attributes(HPLC_HPL$var)$names[i])] <- ncvar_get(HPLC_HPL, attributes(HPLC_HPL$var)$names[i])
}
head(HPLC_HPL_df)
```

```r
# CSV with missing data
HPLC_missing_df <- read.csv("HPLC_Missing.csv")

head(HPLC_missing_df)
```

```r
ChemTax_BBSR$date <- as.POSIXct(as.character(ChemTax_BBSR$Date), format="%Y%m%d")
ChemTax_Mote$date <- as.POSIXct(as.character(ChemTax_Mote$Date), format="%Y%m%d")
ChemTax_HPL$date <- as.POSIXct(as.character(ChemTax_HPL$Date), format="%Y%m%d")
ChemTax_Missing$date <- as.POSIXct(as.character(ChemTax_Missing$Date), format="%Y%m%d")
```

```r
ChemTax_Mote$chl.a[ChemTax_Mote$chl.a=="-9999"] <- "0"
```

```r
ChemTax_HPL$chl.a[as.numeric(ChemTax_HPL$chl.a)>40] <- NA
```

```r
options(repr.plot.width=10, repr.plot.height=8)
a <- ggplot()+
geom_point(data=ChemTax_BBSR,aes(x=date, y=as.numeric(chl.a), col='BBSR'))+
geom_point(data=ChemTax_Mote,aes(x=date, y=as.numeric(chl.a), col='Mote'))+
geom_point(data=ChemTax_Missing,aes(x=date, y=as.numeric(chl.a)/1000, col='missing'))+
scale_x_datetime(date_breaks = "1 year")+
ylab('chl a [µg/l]')+
ggtitle('BBSR + Mote')

b <- ggplot()+
geom_point(data=ChemTax_HPL, aes(x=date, y=as.numeric(chl.a)))+
scale_x_datetime(date_breaks = "1 year")+
ylab('chl a [µg/l]')+
ggtitle('HPL')

plot_grid(a,b, ncol=1)
```
