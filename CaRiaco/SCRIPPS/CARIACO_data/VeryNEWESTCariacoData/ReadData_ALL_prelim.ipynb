```python
import xarray as xr
import pandas as pd
import numpy as np
```

```python
phyto = xr.open_dataset("phytoplankton.netcdf")
```

```python
#phyto
```

```python
phyto['Cruise'] = phyto.Cruise.astype('int')
phyto['Leg'] = phyto.Leg.astype('int')
phyto['Cast'] = phyto.Cast.astype('int')

phyto['Cruise_ID_2'] = phyto.Cruise_ID_2.astype('str')

phyto['Day_local'] = phyto.Day_local.astype('int')
phyto['Month_local'] = phyto.Month_local.astype('int')
phyto['Year_local'] = phyto.Year_local.astype('int')

phyto['Time_start_local'] = phyto.Time_start_local.astype('str')
phyto['Time_end_local'] = phyto.Time_end_local.astype('str')
```

```python
phyto.Datetime_local.values = pd.to_datetime(phyto.Datetime_local.astype('str').values)
phyto.Datetime_UTC.values = pd.to_datetime(phyto.Datetime_UTC.astype('str').values)

phyto.Datetime_local.values = pd.to_datetime(phyto.Datetime_local.astype('str').values)
phyto.Datetime_UTC.values = pd.to_datetime(phyto.Datetime_UTC.astype('str').values)
```

```python
phyto['SpeciesNameOriginal'] = phyto.SpeciesNameOriginal.astype('str')
phyto['SpeciesNameCleaned'] = phyto.SpeciesNameCleaned.astype('str')
phyto['ScientificName_accepted'] = phyto.ScientificName_accepted.astype('str')

phyto['AphiaID'] = phyto.AphiaID.astype('int')
```

```python
def convert_numeric_data(var):
    phyto['d_1m'] = phyto['d_1m'].to_dataset().apply(pd.to_numeric, errors='coerce').to_array()
```

```python
phyto['d_1m'] = phyto.d_1m.astype('float')

phyto['d_7m'] = phyto.d_7m.to_dataset().apply(pd.to_numeric, errors='coerce').to_array()

phyto['d_15m'] = phyto.d_15m.astype('float')
phyto['d_25m'] = phyto.d_25m.astype('float')
phyto['d_35m'] = phyto.d_35m.astype('float')
phyto['d_55m'] = phyto.d_55m.astype('float')

phyto['d_75m'] = phyto.d_75m.to_dataset().apply(pd.to_numeric, errors='coerce').to_array()

phyto['d_100m'] = phyto.d_100m.to_dataset().apply(pd.to_numeric, errors='coerce').to_array()

phyto['Total_sum'] = phyto.Total_sum.astype('float')
```
