```r
ERA5 <- read.csv("ERA5_reanalyisis_monthly.csv")
```

```r
ERA5_clean <- ERA5[complete.cases(ERA5$u100),]
```

```r
ERA5_clean$time <- as.Date(ERA5_clean$time, format="%Y-%m-%d")
```

```r
# Modify This!

plot_var <- function(varnum){
    AirTempDaily <- ggplot(ERA5_clean, aes(time, ERA5_clean[,varnum])) +
           geom_point() +
           ggtitle((names(ERA5_clean)[varnum])) +
           xlab("Date") + ylab("Temperature (C)") +
           scale_x_date(labels=date_format ("%m-%y"))+
           theme(plot.title = element_text(lineheight=.8, face="bold",
                 size = 20)) +
           theme(text = element_text(size=18))

return(AirTempDaily)
    
}
```

```r
plot_var(2)
```
