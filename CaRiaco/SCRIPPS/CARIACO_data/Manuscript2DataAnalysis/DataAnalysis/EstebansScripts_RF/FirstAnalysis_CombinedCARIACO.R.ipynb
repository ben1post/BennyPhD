```r
CARIACO_sub1 <- CARIACO %>%
    select(
u10,
sst,
e,
mtpr,
cdir,

POC_ug_kg,
NO3_merged,
PO4_merged,
SiO4_merged,
Chlorophyll,

total_biomass,
SeaWiFS_chla,
MODIS_chla,
Satellite_chla
          ) %>%
    mutate(irradiance = cdir, wind_speed_u_comp = u10,, 
           SST=sst, evaporation = e, precipiation=mtpr) %>%
select(-u10,
         -cdir,
          -sst,
           
          -e,
          -mtpr,)
```

```r
options(repr.plot.width=30, repr.plot.height=30)
#pdf("Plots/GiantCorrelationPlot.pdf", width=30, height=30) 
pairs(CARIACO_sub1)
#dev.off()
```

```r
cor(CARIACO_sub1, use="complete.obs")
```

```r
cormat <- round(cor(CARIACO_sub1, use="complete.obs"),2)
head(cormat)
```

```r
library(reshape2)
melted_cormat <- melt(cormat)
head(melted_cormat)
```

```r
# Get lower triangle of the correlation matrix
  get_lower_tri<-function(cormat){
    cormat[upper.tri(cormat)] <- NA
    return(cormat)
  }
  # Get upper triangle of the correlation matrix
  get_upper_tri <- function(cormat){
    cormat[lower.tri(cormat)]<- NA
    return(cormat)
  }
```

```r
upper_tri <- get_upper_tri(cormat)
upper_tri
```

```r
options(repr.plot.width=15, repr.plot.height=15)
#library(cowplot)
library(scales)
# Melt the correlation matrix
library(reshape2)
melted_cormat <- melt(upper_tri, na.rm = TRUE)
# Heatmap
library(ggplot2)
ggplot(data = melted_cormat, aes(Var2, Var1, fill = value))+
 geom_tile(color = "white")+
 scale_fill_gradient2(#low = "royalblue3", high ="yellow3", mid = "white", 
   midpoint = 0, limit = c(-1,1), space = "viridis", 
   name="Pearson\nCorrelation") +

  theme_minimal()+ 
 theme(axis.text.x = element_text(angle = 45, vjust = 1, 
    size = 26, hjust = 1), axis.text.y = element_text(size=26))+
 coord_fixed()
```
