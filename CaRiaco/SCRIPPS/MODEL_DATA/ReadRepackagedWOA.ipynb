# NITRATE DATA

```python
WOA_N = xr.open_dataset('REPACKAGED_DATA/Nitrate_WOA_tmld_test03.nc')
```

```python
WOA_N['n0'].units
```

```python
Lat = np.array(WOA_N['lat'])
Lon = np.array(WOA_N['lon'])
n0 = np.array(WOA_N['n0'])

Lats,Lons = np.meshgrid(Lat,Lon,indexing='ij')
```

# NITRATE PLOT

```python
fig = plt.figure(figsize=(18, 16), edgecolor='w')
m = Basemap(projection='cyl',
            llcrnrlat=-90, urcrnrlat=90,
            llcrnrlon=-180, urcrnrlon=180, )
#m.shadedrelief(scale=0.5)

cmap='viridis'

m.pcolormesh(Lons, Lats, n0[:,:,6], latlon=True, cmap=cmap)

#lat = 50
#lon = -70
#x,y = m(lon, lat)
#m.plot(x, y, 'bo', markersize=24)

m.drawcoastlines(color='lightgray')

norm = mpl.colors.Normalize(vmin=np.nanmin(n0), vmax=np.nanmax(n0))
sm = cm.ScalarMappable(norm=norm, cmap=cmap)
#sm.set_array([]) # can be an empty list, only needed for matplotlib < 3.1
# ...
m.colorbar(sm, pad=0.1, label='$N_0$ [$m$ $kg^{-1}$]')
plt.title('Nitrate below mixed layer, calculated from MLD climatology and WOA 2018 - July')
```

```python
fig.savefig('PLOTS_01/01_N0_WOA_July.png')
```

# T_MLD DATA

```python
WOA_t = xr.open_dataset('REPACKAGED_DATA/Temp_WOA_tmld_test02.nc')
```

```python
WOA_t
```

```python
Lat = np.array(WOA_t['lat'])
Lon = np.array(WOA_t['lon'])
tmld = np.array(WOA_t['t_mld'])

Lats,Lons = np.meshgrid(Lat,Lon,indexing='ij')
```

```python
plt.rcParams.update({'font.size': 22})
```

# T_MLD PLOT

```python
def plotglobal(Lons=Lons,Lats=Lats,var=tmld, data='tmld' ,title='Temperature averaged across mixed layer',unit='$T_{MLD}$ [C]' ,month='July'):
    fig = plt.figure(figsize=(18, 16), edgecolor='w')
    m = Basemap(projection='cyl',
                llcrnrlat=-90, urcrnrlat=90,
                llcrnrlon=-180, urcrnrlon=180, )
    #m.shadedrelief(scale=0.5)

    cmap='viridis'
    
    if month == 'January':
        varplot = var[:,:,0]
    elif month == 'July':
        varplot = var[:,:,6]
    else:
        raise('wrong month supplied, try "January" or "July"')

    m.pcolormesh(Lons, Lats, varplot, latlon=True, cmap=cmap)

    #lat = 50
    #lon = -70
    #x,y = m(lon, lat)
    #m.plot(x, y, 'bo', markersize=24)

    m.drawcoastlines(color='lightgray')

    norm = mpl.colors.Normalize(vmin=np.nanmin(var), vmax=np.nanmax(var))
    sm = cm.ScalarMappable(norm=norm, cmap=cmap)
    #sm.set_array([]) # can be an empty list, only needed for matplotlib < 3.1
    # ...
    m.colorbar(sm, pad=0.1, label=unit)
    plt.title(title+ ' - ' + month + '\n $_{calc.}$ $_{from}$ $_{MLD}$ $_{climatology}$ $_{and}$ $_{WOA2018}$')
    
    fig.savefig('PLOTS_01/' + data + '_WOA_' + month +'.pdf')
```

```python
plotglobal(month='January')
```

```python
plotglobal()
```
